EXERCÍCIO 2

```python
def contagem_carac(nome):

  contagem = {}

  for caractere in nome:
    if caractere in contagem:
      contagem[caractere] += 1
    else:
      contagem[caractere] = 1

  return contagem

nome = input("Digite seu nome: ")
print(contagem_carac(nome))
```

EXERCÍCIO 3


```python
def notas_alunos():
    dic_notas = {}
    dic_medias = {}

    n = int(input("quantos alunos deseja cadastrar: "))

    for i in range(n):
        nome = input("digite seu nome: ")
        n1 = float(input("Digite a nota 1: "))
        n2 = float(input("Digite a nota 2: "))
        #criar uma lista dentro do dicionário
        notas = []
        notas.append(n1)
        notas.append(n2)
        dic_notas.update({nome:notas})


        medias = []
        media = (n1 + n2)/2
        medias.append("%.2f"% media)
        dic_medias.update({nome:medias})

    print("Lista de alunos e suas notas: ", dic_notas)
    print("Lista de alunos e suas médias: ", dic_medias)

    return

notas_alunos()
```

```python
def retornar_lista():
  lista = []

  for i in range(5):
    numero = int(input("Digite um número: "))
    lista.append(numero)

  return lista

resultado = retornar_lista()
print("Lista de números digitados: ", resultado)
```

```python
def somar_elementos():

  lista = []

  for i in range(5):
    numero = int(input("Digite um número: "))
    lista.append(numero)

  soma = sum(lista)

  return soma

print("Soma dos números digitados: ", somar_elementos())
```
