EXERCÍCIO 1

```python
class Fila:
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.inicio = 0
    self.final = -1
    self.elementos = 0
    self.valores = [''] * capacidade

  def fila_cheia(self):
    return self.elementos == self.capacidade

  def enfileirar(self, valor):
    if self.fila_cheia():
      print("fila cheia")
      return
    elif self.final == self.capacidade - 1:
        self.final = -1
    self.final += 1
    self.valores[self.final] = valor
    self.elementos += 1

  def fila_vazia(self):
    return self.elementos == 0

  def desenfileirar(self):
    if self.fila_vazia():
      print("fila vazia")
      return
    temp = self.valores[self.inicio]
    self.inicio += 1
    if self.inicio == self.capacidade:
      self.inicio = 0
    self.elementos -= 1
    return temp

  def primeiro_fila(self):
    if self.fila_vazia():
      return -1
    return self.valores[self.inicio]

fila = Fila(5)

fila.desenfileirar() #fila vazia

fila.enfileirar('j')
fila.enfileirar('u')
fila.enfileirar('l')
fila.enfileirar('i')
fila.enfileirar('a')

fila.enfileirar('a') #fila cheia

mostrar_primeiro = fila.primeiro_fila()
print(mostrar_primeiro)

fila.desenfileirar()
fila.desenfileirar()
fila.desenfileirar()

fila.primeiro_fila() #primeiro depois de desenfileirar
```

```python
class Pilha:
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.topo = -1
    self.valores = [''] * capacidade

  def empilhar(self, valor):
      print (self.topo)
      if self.topo == self.capacidade - 1:
        print("pilha cheia!")
      else:
        self.topo += 1
        self.valores[self.topo] = valor

  def ver_topo(self):
      if self.topo == -1:
        print("pilha vazia!")
      else:
        return self.valores[self.topo]

  def pilha_vazia(self):
      if self.topo == -1:
        print("pilha vazia!")
        return

  def desempilhar(self):
      if self.pilha_vazia():
        return
      else:
        self.topo -= 1

pilha = Pilha(5)

pilha.desempilhar()

pilha.empilhar('j')
pilha.empilhar('u')
pilha.empilhar('l')
pilha.empilhar('i')
pilha.empilhar('a')
pilha.empilhar('a')

pilha.ver_topo()

pilha.desempilhar()
pilha.desempilhar()
pilha.desempilhar()
pilha.desempilhar()

pilha.ver_topo()
```
