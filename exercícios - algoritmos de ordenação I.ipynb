EXERCÍCIO 1

```python
def bubble_sort(lista):
  n = len(lista)
  for i in range(n):
    for j in range(0, n - i - 1):
      if lista[j] > lista[j + 1]:
        temp = lista[j]
        lista[j] = lista[j + 1]
        lista[j + 1] = temp
  return lista

import random

lista_num = [random.randint(0, 10000) for _ in range(1000)]
bubble_sort(lista_num)

print("lista ordenada de números: ", lista_num)

lista_nomes = ['ana', 'nikolas', 'thyerri', 'lara', 'cida', 'maria']
bubble_sort(lista_nomes)

print("lista ordenada de nomes: ", lista_nomes)
```

EXERCÍCIO 2

```python
def insertion_sort(lista):
  n = len(lista)
  for i in range(1, n):
    marcador = lista[i]
    j = i - 1
    while j >= 0 and marcador < lista[j]:
      lista[j + 1] = lista[j]
      j = j - 1
    lista[j + 1] = marcador
  return lista

import random

lista_num = [random.randint(0, 10000) for _ in range(1000)]
insertion_sort(lista_num)

print("lista ordenada de números: ", lista_num)

lista_nomes = ['julia', 'nikolas', 'beatriz', 'lara', 'giovana', 'miguel']
insertion_sort(lista_nomes)

print("lista ordenada de nomes: ", lista_nomes)
```
