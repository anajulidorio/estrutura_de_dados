```python
def contar_digitos(numero):
    if numero < 10:
      return 1
    else:
      return 1 + contar_digitos(numero//10)

print(contar_digitos(12345))
```

```python
def soma_lista(lista):
  if len(lista) == 0:
    return 0
  else:
    return lista[0] + soma_lista(lista[1:])


print(soma_lista([1, 2, 3, 4, 5]))
```

```python
def verificar_palindromo(palavra):
  if len(palavra) == 1:
    return True
  if palavra[0] != palavra[-1]:
    return False
  return verificar_palindromo(palavra[1:-1])

print("radar: ", verificar_palindromo("radar"))
print("python: ", verificar_palindromo("python"))
print("ana: ", verificar_palindromo("ana"))
```

```python
def desenhar_triangulo(n, linha = 1):
  if linha > n:
    return
  print("*" * linha)
  desenhar_triangulo(n, linha + 1)

desenhar_triangulo(5)
```
